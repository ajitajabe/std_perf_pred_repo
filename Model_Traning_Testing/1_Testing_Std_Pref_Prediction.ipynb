```python
def get_std_perf_pre(Gender, AttendanceRate,StudyHoursPerWeek,PreviousGrade,ExtracurricularActivities,ParentalSupport):

    Gender=Columns_data["Gender"][Gender]
    AttendanceRate=np.sqrt(AttendanceRate)
    ParentalSupport=Columns_data["ParentalSupport"][ParentalSupport]
    test_array=np.zeros((1,feature_count))
    test_array[0,0]=Gender
    test_array[0,1]=AttendanceRate
    test_array[0,2]=StudyHoursPerWeek
    test_array[0,3]=PreviousGrade
    test_array[0,4]=ExtracurricularActivities
    test_array[0,5]=ParentalSupport
    Std_pred_per=np.around(Std_Per_Pred_model.predict(test_array)[0])
    return Std_pred_per
```

```python
Gender= "Female"
AttendanceRate=81
StudyHoursPerWeek=21
PreviousGrade=85
ExtracurricularActivities=2
ParentalSupport="High"
get_std_perf_pre(Gender, AttendanceRate,StudyHoursPerWeek,PreviousGrade,ExtracurricularActivities,ParentalSupport)
```
